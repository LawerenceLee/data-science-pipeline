# Visualizing Data 

> Visualizations complement summary statistics really well and give you a broad purview of your data.

Matplotlib is the library that deals with the visualization. Enforcing inline will output our visualizations to the cells of the Jupyter Notebook.

```python
import pandas as pd 
import numpy as np 
from pandas import Series, DataFrame
import matplotlib.pyplot as plt       
%matplotlib inline                    
```

```python
file_path = './data_files/dataset_Facebook.csv' # specify the file_path
df = pd.read_csv(file_path, delimiter=';') # specify the delimiter as semi-colon and read the file via read_csv method
```

# Distribution

### Histogram

```python
df.hist(figsize=(15,40),layout=(10,2),sharex=False, sharey=False) # use the hist method to create a histogram of each feature
plt.show()
```

Remember Post Weekday was the feature that had nearly zero skew. And unsuprisingly the visualization shows this. It looks nearly uniform.

### Alternative 1 to Histogram: Density Plot

```python
# Think of this as a smoothened out histogram
df.plot(kind='density',figsize=(15,40), layout=(10,2),subplots=True,sharex=False, sharey=False)
plt.show()
```

### Alternative 2 to Histogram: Box and Whisker Plot

```python
# the box shows the mid 50% of the data 
df.plot(kind='box',figsize=(15,40),subplots=True,layout=(10,2),sharex=False, sharey=False)
plt.show()
```

# Correlation

> Before we did df.corr() to get a correlation matrix, here we will visualize correlation.

```python
df.corr(method='pearson')
```

It's actually going to be pretty unwieldy if we try to visualize a correlation matrix of this size. Let's reduce the scope to 5 features.

```python
few_columns = df.columns[:6].tolist()
```

```python
few_columns
```

```python
correlations = df[few_columns].corr(method='pearson')
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(correlations,vmin=-1,vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,5,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(['']*len(few_columns))
ax.set_yticklabels(few_columns)
plt.tight_layout()
plt.show()
```

### Scatter Plot Matrix

```python
# this is a really cool tool that essentially plots a scatter plot showing all of the correlations
from pandas.tools.plotting import scatter_matrix
```

Because there is little
point of drawing a scatter plot of each variable with itself, the diagonal shows histograms of
each attribute.

Again this scatter matrix is unwieldy with so many features to visualize so we're restricting to visualizing some of the features.

```python
scatter_matrix(df[few_columns], alpha=0.1, figsize=(10,10), diagonal='kde')
plt.show()
```
